# Import Library

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

# Read and Exploration Data

```python
df = pd.read_csv("https://raw.githubusercontent.com/Prenagen24/Country-Demographic-Visualisation/main/factbook.csv")
df
```

# Task 1

### Visualisation 1

```python
#Assign ke Dataframe Baru
df_vis1 = df[['Continent','GDP','Life expectancy at birth','Population']]
df_vis1
```

```python
#Plot
sns.set_theme()
sns.set(rc={"figure.figsize":(24, 16)})
cmap = sns.color_palette("Set2",n_colors=5)
g = sns.relplot(
    data=df_vis1,
    x="Population", y="GDP",
    hue="Continent", size="Life expectancy at birth",
    palette=cmap,sizes=(10,200)
)
g.set(xscale="log", yscale="log",title='Life Expectancy at Brith per Continent based on GDP and Population')
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

# Task 2

### Visualisation 1

```python
#Menghitung Trade Balance
trade_balance = df['Exports']-df['Imports']
trade_balance
```

```python
#Assign ke Dataframe Baru
df_vis3 = pd.DataFrame()
df_vis3['Continent'] = df['Continent']
df_vis3['Trade Balance'] = trade_balance
df_vis3
```

```python
#Plot
tmpdf_vis3 = df_vis3.groupby(['Continent']).mean().plot(kind='bar',figsize=(10,5),title='Average Trade Balance per Continent')
```

### Visualisation 2

```python
# Menghitung selisih rata-rata Oil Consumption antara Asia dan benua lainnya
asia_avg = df[df['Continent'] == 'Asia']['Oil consumption'].mean()
other_continents_avg = df[df['Continent'] != 'Asia']['Oil consumption'].mean()
diff_avg = asia_avg - other_continents_avg
```

```python
# Membuat bar chart
plt.figure(figsize=(10,5))
plt.bar(['Asia', 'Other Continents'], [asia_avg, other_continents_avg])
plt.title('Average Oil consumption: Asia vs Other Continents')
plt.xlabel('Continent')
plt.ylabel('Average Oil consumption')
plt.show()
```

### Visualisation 3

```python
def divide_two_cols(df_sub):
    df_sub['divs'] = df_sub['Infant mortality rate'] / float(df_sub['Birth rate'].sum())
    return df_sub

dfinfant = df[['Infant mortality rate','Birth rate','Continent']]
dfbaru = dfinfant.groupby('Continent').apply(divide_two_cols)
dfbaru = dfbaru.groupby('Continent').mean()['divs']
dfbaru = dfbaru.reset_index()
dfbaru = dfbaru.drop(4)
dfbaru = dfbaru.groupby('Continent').mean()['divs'].plot(kind='bar',figsize=(10,5),title='Average Infant death per birth rate each continent')
```
